Business operations. IT leverages Incident Management process to achieve the above Objective. The main goal of Incident Management process is to provide a quick fix / workarounds or solutions that resolves the 
interruption and restores the service to its full capacity to ensure no business impact. In most of the organizations, incidents are created by various Business and IT Users, End Users/ Vendors if they have access to ticketing systems, and from the integrated monitoring systems and tools. Assigning the incidents to the appropriate person or unit in the support team has critical importance to provide improved user satisfaction while ensuring better allocation of support resources. The assignment of incidents to appropriate IT groups is still a manual process in many of the IT organizations. Manual assignment of incidents is time consuming and requires human efforts. There may be mistakes due to human errors and resource consumption is carried out ineffectively because of the misaddressing. On the other hand, manual assignment increases the response and resolution times 
which result in user satisfaction deterioration / poor customer service.

Around ~54% of the incidents are resolved by L1 / L2 teams. Incase L1 / L2 is unable to resolve, they will then escalate / assign the tickets to Functional teams from Applications and Infrastructure (L3 teams). <span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">2a.Text case[upper or lower] handling


```python
#Text case [upper or lower] handling
def recursion_lower(x):
    if type(x) is str:
        return x.lower()
    elif type(x) is list:
        return [recursion_lower(i) for i in x]
    elif type(x) is dict:
        return {recursion_lower(k):recursion_lower(v) for k,v in x.items()}
    else:
        return x
```

```python
Corpus=recursion_lower(Corpus)
print(Corpus)
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">2b.Tokenisation

```python
#Download "punkt" if missing
#nltk.download('punkt')
```

```python
#Tokenisation

#Extract data
W=[]#Tokens
L=[]#Identified Tags of Labels
doc_x=[]#Tokenised words
doc_y=[]#Tags or Labels


for intent in Corpus['intents']:
        for pattern in intent['patterns']:
            w_temp=nltk.word_tokenize(pattern)
            W.extend(w_temp)
            doc_x.append(w_temp)
            doc_y.append(intent["tag"])

    #Add the missing tag if any
        if intent['tag'] not in L:
            L.append(intent['tag'])
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">2c.Stemming


```python
#Stemming
#It is a process to find similarities between words with the same root words.
#This will help us to reduce the bag of words by associating similar words with their corresponding root words.

from nltk.stem import LancasterStemmer
#from nltk.stem import WordNetLemmatizer
stemmer = LancasterStemmer()
#stemmer=WordNetLemmatizer()

W=[stemmer.stem(w.lower()) for w in W if w!="?"]#Stemming or learning the root word
W=sorted(list(set(W)))#Sorted words
L=sorted(L)#Sorted list of tags or label
```

```python
word_counts = dict()

for i in W:
        if i in word_counts:
            word_counts[i] += 1
        else:
            word_counts[i] = 1
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">2d.Generate BOW [Bag of Words]


Generate BOW [Bag of Words]

Process of converting words into numbers by generating vector embeddings from the tokens generated above.
This is given as an input to the neural network model for understanding the written text.

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">3.Generate one hot encoding for the target column

One hot encode the output or targets [In our case we have defined them as “TAG” in the corpus]

Process of converting words into numbers by generating vector embeddings from the tokens generated above.


```python
Train=[]#Training data for NN
Target=[]#Target data for NN

out_empty=[0 for _ in range(len(L))]


#Loop to create bag of words and put the frequency count on each word
for x,doc in enumerate(doc_x):
    bag=[]

    w_temp=[stemmer.stem(w.lower()) for w in doc]

    for w in W:
            if w in w_temp:
                bag.append(1)
            else:
                bag.append(0)


    output_row=out_empty[:]
    output_row[L.index(doc_y[x])]=1

    Train.append(bag) #List
    Target.append(output_row) #List
```

```python
import numpy as np
# Finally we will convert our training data and output to numpy arrays   
Train =np.array(Train)       
Target =np.array(Target)
```

<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#00b3e5;">4.Design a neural network to classify the words with TAGS as target outputs

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras import losses
from tensorflow.keras import regularizers,optimizers
```

```python
#Keras start up
model=Sequential()
#NN Layers
model.add(Dense(16,input_dim=len(Train[0]),activation='relu'))#i/p layer
model.add(Dropout(0.5))
model.add(Dense(16,activation='relu'))#hidden layer
model.add(Dropout(0.5))
model.add(Dense(len(Target[0]),activation='softmax'))#O/p layer
```

```python
model.summary()
```

```python
# Compile the model - add mean squared error as loss and stochastic gradient descent as optimizer
from tensorflow.keras import optimizers
opt = optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=opt,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model_history=model.fit(Train, Target,epochs=200,batch_size=8,verbose=2)
```

```python
def bag_of_words(s, words):
    bag = [0 for _ in range(len(words))]

    s_words = nltk.word_tokenize(s)
    s_words = [stemmer.stem(word.lower()) for word in s_words]
    
    for se in s_words:
        for i, w in enumerate(words):
            if w == se:
                bag[i] = 1
    
    return np.array(bag)
```

5.Design a chat utility as a function to interact with the user till the user calls a “quit”
        
6.If the user does not understand or finds the bot’s answer irrelevant, the user calls a “*” asking the bot to re-evaluate what the user has asked
    

```python
#Text chat utility function

def chat():
        print("Chat (type:stop to quit)")
        print("If answer is not right(type:*)")
        while True:
            inp=input("\n\nYou:")
            
            if inp.lower()=="*":
                print("BOT:Please rephrase your question and try again")
            if inp.lower()=="quit":
                break
         
            results=model.predict(np.asanyarray([bag_of_words(inp,W)]))[0]

            results_index=np.argmax(results)
            tag=L[results_index]


            for tg in Corpus["intents"]:
                if tg['tag']==tag:
                    responses=tg['responses']

            print(random.choice(responses))
```
